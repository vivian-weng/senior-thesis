```python
# video_path = "/content/drive/My Drive/Colab Notebooks/CPSC 490/FR1/Viv_free_side.mp4"
# output_dir = "/content/drive/My Drive/Colab Notebooks/CPSC 490/FR1/"

# csv_file = "/content/drive/My Drive/Colab Notebooks/CPSC 490/FR1/Viv_free_side.csv"

"""Function divides a video into frames, placing frames into
folders with 30 frames each and a spreadsheet containing
a column of the frame file names and a column that corresponds
to the stroke classification label"""
def parse_video(origin_folder, video_name, stroke):
  # Stroke classification labels
  stroke_dict = {
      "FL": 0,
      "BK": 1,
      "BR": 2,
      "FR": 3
  }

# Setting video path and cap
  video_path = origin_folder + video_name + ".mp4"
  print(video_path)
  cap = cv2.VideoCapture(video_path)

  frame_count = 0
  # opening cap
  while(cap.isOpened()):
    output_dir = origin_folder + f"Segment{frame_count // 30}/"

    if not os.path.exists(output_dir):
      os.makedirs(output_dir)
    ret, frame = cap.read()


    if ret == True:
      # Create frame name
      frame_name = f'frame_{frame_count:04d}.jpg'
      # Place frame into output directory
      cv2.imwrite(output_dir + frame_name, frame)

      # Print output checkpoint every 100 frames
      if frame_count % 100 == 0:
        print(f"Frame {frame_count} extracted successfully for {video_name}")

      frame_count += 1

      # Create a csv file in same folder
      csv_file = output_dir + f"labels.csv"
      # Write frame name and stroke classification label to csv file
      with open(csv_file, 'a+') as file:
        file.write(f'{frame_name}, {stroke_dict[stroke]} \n')
        file.close()

    else:
      break

  # Release cap
  cap.release()
  cv2.destroyAllWindows
  print(f'Frame extracted: {frame_count}')

```
